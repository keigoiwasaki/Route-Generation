```python
import folium
from folium import plugins
from collections import defaultdict
import json
import pandas as pd
from shapely.geometry import shape, LineString

def read_geojson(route_path):
    """
    GeoJSONを読み込み、Pandas DataFrameに変換する
    """
    with open(route_path, "r", encoding="utf-8") as f:
        geojson_data = json.load(f)

    # 属性データとジオメトリを抽出
    data = []
    for feature in geojson_data["features"]:
        properties = feature["properties"]  # 属性データ
        geometry = shape(feature["geometry"])  # Shapelyジオメトリに変換
        properties["geometry"] = geometry
        data.append(properties)

    # Pandas DataFrameに変換
    df = pd.DataFrame(data)
    return df

# GeoJSONファイルのパスを設定
geojson_file_path = "data/bus_route.geojson"  # 実際のファイルパスを設定

# GeoJSONデータをDataFrameに変換
df = read_geojson(geojson_file_path)

# 地図の初期設定 (白ベースのタイルを設定)
m = folium.Map(
    location=[33.96258, 131.2453132],
    zoom_start=15,
    tiles="CartoDB positron"  # 白ベースの地図
)

# 座標ごとに路線を集約する辞書
coordinate_routes = defaultdict(list)

# DataFrameを解析し、座標の重複をチェック
for _, row in df.iterrows():
    route_name = f"{row['from_stop_name']} -> {row['to_stop_name']}"
    geometry = row["geometry"]
    
    # LineString の各座標ペア (セグメント) を取得
    coords = list(geometry.coords)
    for i in range(len(coords) - 1):
        segment = (coords[i], coords[i + 1])
        coordinate_routes[segment].append(route_name)
        # 逆方向も重複とみなす
        reverse_segment = (coords[i + 1], coords[i])
        coordinate_routes[reverse_segment].append(route_name)

# レイヤーの初期化
outlines = folium.FeatureGroup("outlines")
line_bg = folium.FeatureGroup("lineBg")
bus_lines = folium.FeatureGroup("busLines")

# ラインの太さと色を設定
line_weight = 2  # 細めのライン
line_color = "green"  # 緑色のバス路線

# DataFrameの各路線をFoliumで表示
for _, row in df.iterrows():
    route_name = f"{row['from_stop_name']} -> {row['to_stop_name']}"
    geometry = row["geometry"]
    coords = list(geometry.coords)
    
    # 各セグメントの重複状態をチェック
    for i in range(len(coords) - 1):
        segment = (tuple(coords[i]), tuple(coords[i + 1]))
        overlapping_routes = coordinate_routes[segment]
        
        segment_coords = [[coord[1], coord[0]] for coord in [segment[0], segment[1]]]
        segment_width = len(overlapping_routes) * (line_weight + 1)

        # 黒と白のアウトラインを追加 (白い帯効果)
        folium.PolyLine(
            segment_coords,
            color="#000",  # 黒色の外枠
            weight=segment_width + 6,  # 幅を少し広げる
            opacity=1,
        ).add_to(outlines)

        folium.PolyLine(
            segment_coords,
            color="#fff",  # 白色の内側
            weight=segment_width + 4,  # メインのラインより太く
            opacity=1,
        ).add_to(line_bg)

        # 重複する路線がある場合も同じ色・同じ太さでオフセット表示
        if len(overlapping_routes) > 1:
            for j, other_route in enumerate(overlapping_routes):
                plugins.PolyLineOffset(
                    segment_coords,
                    color=line_color,
                    weight=line_weight,
                    opacity=1,
                    offset=j * (line_weight + 1) - (segment_width / 2) + ((line_weight + 1) / 2),
                    tooltip=other_route,
                ).add_to(bus_lines)
        else:
            # 重複しない場合も同じ緑色で表示
            folium.PolyLine(
                segment_coords,
                color=line_color,
                weight=line_weight,
                opacity=1,
                tooltip=route_name,
            ).add_to(bus_lines)

# レイヤーを地図に追加
outlines.add_to(m)
line_bg.add_to(m)
bus_lines.add_to(m)

# 地図を保存または表示
output_file = "green_routes_with_white_outline_map.html"
# m.save(output_file)
print(f"地図を保存しました: {output_file}")
```

```python
# GeoJSONファイルのパスを設定
geojson_file_path = "data/bus_route.geojson"

# GeoJSONデータをDataFrameに変換
with open(geojson_file_path, "r", encoding="utf-8") as f:
    geojson_data = json.load(f)

data = []
for feature in geojson_data["features"]:
    properties = feature["properties"]
    geometry = shape(feature["geometry"])
    properties["geometry"] = geometry
    data.append(properties)

df = pd.DataFrame(data)

# 地図の初期設定 (白ベースのタイルを設定)
m = folium.Map(
    location=[33.96258, 131.2453132],
    zoom_start=15,
    tiles="CartoDB positron"
)

# 座標ごとに路線を集約する辞書 (正方向と逆方向を分けて保持)
coordinate_routes = defaultdict(lambda: {"forward": set(), "reverse": set()})

# DataFrameを解析し、座標の重複をチェック
for _, row in df.iterrows():
    route_name = f"{row['from_stop_name']} → {row['to_stop_name']}"
    geometry = row["geometry"]
    coords = list(geometry.coords)

    for i in range(len(coords) - 1):
        segment = (coords[i], coords[i + 1])
        reverse_segment = (coords[i + 1], coords[i])
        
        # 正方向は "forward"、逆方向は "reverse" に追加 (重複を避ける)
        coordinate_routes[segment]["forward"].add(route_name)
        coordinate_routes[reverse_segment]["reverse"].add(route_name)

# レイヤーの初期化
outlines = folium.FeatureGroup("Outlines")
line_bg = folium.FeatureGroup("Line Background")
bus_lines = folium.FeatureGroup("Bus Lines")

# ラインの設定
line_weight = 3  # 線の太さ (より細く調整)
line_color = "green"  # 線の色 (緑)

# 特定の条件に合致する座標ペアのみを表示
for segment, directions in coordinate_routes.items():
    forward_routes = list(directions["forward"])
    reverse_routes = list(directions["reverse"])
    
    segment_coords = [[coord[1], coord[0]] for coord in [segment[0], segment[1]]]

    # 白いアウトラインを追加 (白い帯効果)
    folium.PolyLine(
        segment_coords,
        color="#fff",
        weight=line_weight * 2 + 2,
        opacity=1,
    ).add_to(outlines)

    folium.PolyLine(
        segment_coords,
        color="#fff",
        weight=line_weight * 2,
        opacity=1,
    ).add_to(line_bg)

    # 1. 正方向と逆方向の両方が存在する場合 (2本表示)
    if forward_routes and reverse_routes:
        
        # 逆方向 (より近づけたオフセットを適用)
        plugins.PolyLineOffset(
            segment_coords,
            color=line_color,
            weight=line_weight,
            opacity=2,
            offset=2,  # オフセットを最小限に設定
            tooltip=reverse_routes[0] if reverse_routes else "逆方向",
        ).add_to(bus_lines)

    # 2. 正方向のみ存在する場合 (1本表示)
    elif forward_routes:
        folium.PolyLine(
            segment_coords,
            color=line_color,
            weight=line_weight,
            opacity=1,
            tooltip=forward_routes[0],
        ).add_to(bus_lines)

    # 3. 逆方向のみ存在する場合 (1本表示)
    elif reverse_routes:
        folium.PolyLine(
            segment_coords,
            color=line_color,
            weight=line_weight,
            opacity=1,
            tooltip=reverse_routes[0],
        ).add_to(bus_lines)

# レイヤーを地図に追加
outlines.add_to(m)
line_bg.add_to(m)
bus_lines.add_to(m)

# 地図を保存
output_file = "green_routes_with_non_overlapping_map.html"
# m.save(output_file)
print(f"地図を保存しました: {output_file}")
```

```python
import folium
from folium import plugins
from collections import defaultdict
import json
import pandas as pd
from shapely.geometry import shape
from itertools import cycle

# GeoJSONファイルのパスを設定
geojson_file_path = "data/bus_route.geojson"

# GeoJSONデータをDataFrameに変換
with open(geojson_file_path, "r", encoding="utf-8") as f:
    geojson_data = json.load(f)

data = []
for feature in geojson_data["features"]:
    properties = feature["properties"]
    geometry = shape(feature["geometry"])
    properties["geometry"] = geometry
    data.append(properties)

df = pd.DataFrame(data)

# 地図の初期設定 (白ベースのタイルを設定)
m = folium.Map(
    location=[33.96258, 131.2453132],
    zoom_start=15,
    tiles="CartoDB positron"
)

# 座標ごとに路線を集約する辞書 (正方向と逆方向を分けて保持)
coordinate_routes = defaultdict(lambda: {"forward": set(), "reverse": set()})

# DataFrameを解析し、座標の重複をチェック
for _, row in df.iterrows():
    route_name = f"{row['from_stop_name']} → {row['to_stop_name']}"
    geometry = row["geometry"]
    coords = list(geometry.coords)

    for i in range(len(coords) - 1):
        segment = (coords[i], coords[i + 1])
        reverse_segment = (coords[i + 1], coords[i])
        
        # 正方向は "forward"、逆方向は "reverse" に追加 (重複を避ける)
        coordinate_routes[segment]["forward"].add(route_name)
        coordinate_routes[reverse_segment]["reverse"].add(route_name)

# 5色のカラーパレットを設定
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]
color_cycle = cycle(colors)  # 色を循環して使う

# ルート名ごとに色を割り当て (一貫性を保つ)
route_colors = {}
for routes in coordinate_routes.values():
    for route_name in routes["forward"] | routes["reverse"]:
        if route_name not in route_colors:
            route_colors[route_name] = next(color_cycle)

# レイヤーの初期化
outlines = folium.FeatureGroup("Outlines")
line_bg = folium.FeatureGroup("Line Background")
bus_lines = folium.FeatureGroup("Bus Lines")

# ラインの設定
line_weight = 3  # 線の太さ
offset_distance = 4  # オフセット距離

# 特定の条件に合致する座標ペアのみを表示
for segment, directions in coordinate_routes.items():
    forward_routes = list(directions["forward"])
    reverse_routes = list(directions["reverse"])
    
    segment_coords = [[coord[1], coord[0]] for coord in [segment[0], segment[1]]]

    # 白いアウトラインを追加 (白い帯効果)
    folium.PolyLine(
        segment_coords,
        color="#fff",
        weight=line_weight * 2 + 2,
        opacity=1,
        highlight=False
    ).add_to(outlines)

    folium.PolyLine(
        segment_coords,
        color="#fff",
        weight=line_weight * 2,
        opacity=1,
        highlight=False
    ).add_to(line_bg)

    # 1. 正方向と逆方向の両方が存在する場合 (2本表示)
    if forward_routes and reverse_routes:
        route_name = forward_routes[0] if forward_routes else reverse_routes[0]
        route_color = route_colors[route_name]

        # 逆方向 (オフセットを適用)
        plugins.PolyLineOffset(
            segment_coords,
            color=route_color,
            weight=line_weight,
            opacity=1,
            offset=offset_distance,
            tooltip=reverse_routes[0] if reverse_routes else "逆方向",
            highlight_options={"weight": 0, "color": None, "opacity": 0}  # エッジ表示を無効化
        ).add_to(bus_lines)

    # 2. 正方向のみ存在する場合 (1本表示)
    elif forward_routes:
        route_color = route_colors[forward_routes[0]]
        folium.PolyLine(
            segment_coords,
            color=route_color,
            weight=line_weight,
            opacity=1,
            tooltip=forward_routes[0],
            highlight_options={"weight": 0, "color": None, "opacity": 0}  # エッジ表示を無効化
        ).add_to(bus_lines)

    # 3. 逆方向のみ存在する場合 (1本表示)
    elif reverse_routes:
        route_color = route_colors[reverse_routes[0]]
        folium.PolyLine(
            segment_coords,
            color=route_color,
            weight=line_weight,
            opacity=1,
            tooltip=reverse_routes[0],
            highlight_options={"weight": 0, "color": None, "opacity": 0}  # エッジ表示を無効化
        ).add_to(bus_lines)

# レイヤーを地図に追加
outlines.add_to(m)
line_bg.add_to(m)
bus_lines.add_to(m)

# 地図を保存
output_file = "consistent_colored_routes_no_edge_map.html"
# m.save(output_file)
print(f"地図を保存しました: {output_file}")
```
